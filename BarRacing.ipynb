# Data

```python
df = pd.read_csv('Installs.csv')
```

# Data Prep

```python
df["Date Group"] = pd.to_datetime(df["Date Group"])
```

## Cumulative Sums

```python
cumulative_sums = df.drop("Date Group", axis=1).cumsum()

# "Date Group" column for reference:
cumulative_sums["Date Group"] = df["Date Group"]

# Set "Date Group" as the index
cumulative_sums.set_index("Date Group", inplace=True)
```

```python
cumulative_sums
```

# Racing Bar Chart

## Installs

```python
def total_installs_period(v, r):
    total = v.sum()
    return {'s': f'Total installs: {total:,.0f}', 'x': .95, 'y': .05, 'ha': 'right', 'size': 8}

html_content = bcr.bar_chart_race(
    df=cumulative_sums,
    filename=None, #change this to save as a local .mp4 file
    title='Cumulative Installs Per Network Over Time',
    period_summary_func=total_installs_period,
    period_length=500  
)

display(html_content)
```
